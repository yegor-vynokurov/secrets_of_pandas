# How to set a two y axis and major and minor grid

```python
n = 700
fig, ax1 = plt.subplots(figsize = (20, 10))

ax2 = ax1.twinx()

ax1.plot(df.loc[n:, 'Open'], label = 'Open', color = 'orange')
ax1.plot(df.loc[n:, 'Close'], label = 'Close', color = 'red')
ax2.plot(df.loc[n:, 'Close_dif_1'], label = 'First difference of Close', color = 'green')
ax2.plot(df.loc[n:, 'Open_dif_1'], label = 'First difference of Open', color = 'blue')

ax1.set_ylabel('Деньги (тёплые цвета)')
ax2.set_ylabel('Разность денег (холодные цвета)')

ax1.legend(loc = 'upper right')
ax2.legend(loc = 'upper center')

plt.minorticks_on()
plt.grid(which='major', linestyle = '-')
plt.grid(which='minor', linestyle = ':')

plt.show()
```

# Frequency of xticks on x axis in matplotlib

## Frequency of dates and rotation of xtick's names

```python
df['race_date_2'] = pd.to_datetime(df['race_date'])
```

```python
plt.figure(figsize = (25,8))
plt.bar(df.race_date_2.value_counts().index, df.race_date_2.value_counts())
plt.xticks(np.arange(df.race_date_2.value_counts().index.min(), # from min 
                     df.race_date_2.value_counts().index.max() + pd.Timedelta(5, unit='D'), # to max + reserve of some days
                     pd.Timedelta(5, unit='D')), rotation = 90)
plt.grid()
plt.show()
```

## Frequency of values

```python
plt.figure(figsize = (25,8))
sns.histplot(df.run_up_distance)
plt.xticks(np.arange(0,df.run_up_distance.max(), 5), rotation = 45)
plt.show()
```
