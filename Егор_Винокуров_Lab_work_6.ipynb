# Візуалізація, як в PowerBI

```python
df = pd.read_csv('/content/drive/MyDrive/Classroom/Data Science Foundation - 14/basa_all_num_sorted5 (1).csv', index_col=0)
df.index = pd.to_datetime(df.index)
```

```python
sns.set_style("whitegrid")
fig, axes = plt.subplots(figsize=(15,9))
position = range(1,5)
sns.lineplot(x = df.index.quarter.values, y = df['Price'].values, hue = df['goal'].values, ci = 0)
axes.set_xticks(position)

axes.set_xticklabels(['Перший квартал', 'Другий квартал', 'Третій квартал', 'Четвертий квартал'])
# axes.set_xticklabels(['', 'Перший квартал', '', 'Другий квартал', '', 'Третій квартал', '', 'Четвертий квартал'])
axes.set_ylabel('Ціна')
axes.set_xlabel('Квартал')
axes.set_title('Залежність ціни від кварталу')
plt.show()
```

```python
df.groupby('Currency')['Currency'].count()
```

```python
sns.set_style("whitegrid")
fig, axes = plt.subplots(figsize=(15,9))

plt.pie(df.groupby('Currency')['Currency'].count(), labels = df.groupby('Currency')['Currency'].count().index, 
        autopct='%.2f%%', wedgeprops=dict(width=0.3))
# my_circle=plt.Circle( (0,0), 0.7, color='white')
# p=plt.gcf()
# p.gca().add_artist(my_circle)

axes.set_title('Розподіл валют операцій')
plt.show()
```

```python
df.groupby('Name')['Name'].count()
```

```python
sns.set_style("whitegrid")
fig, axes = plt.subplots(figsize=(10, 70))
plots = sns.barplot(x = df.groupby('Name')['Name'].count(), y = df.groupby('Name')['Name'].count().index,
            ci=95, orient = 'h', 
            )

plt.xlabel('Кількість')
plt.ylabel('Назва цінного паперу')
plt.title('Розподіл кількості цінних паперів за назвою')
plt.show()
```

```python
df.groupby('goal')['Value'].mean()
```

```python
sns.set_style("whitegrid")
fig, axes = plt.subplots(figsize=(15, 9))
plots = sns.barplot(y = df.groupby('goal')['Value'].mean(), x = df.groupby('goal')['Value'].mean().index,
            ci=95, orient = 'v'
            )
for bar in plots.patches:
   
  # Using Matplotlib's annotate function and
  # passing the coordinates where the annotation shall be done
  # x-coordinate: bar.get_x() + bar.get_width() / 2
  # y-coordinate: bar.get_height()
  # free space to be left to make graph pleasing: (0, 8)
  # ha and va stand for the horizontal and vertical alignment
    plots.annotate(format(bar.get_height(), '.2f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=15, xytext=(0, 8),
                   textcoords='offset points')
plt.xlabel('Цільовий результат')
plt.ylabel('Середня вартість цільового результату')
plt.title('Розподіл середньої вартості в залежності від цільового результату')
plt.show()
```
